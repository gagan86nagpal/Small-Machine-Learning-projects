```python
%matplotlib notebook
import pandas as pd
import numpy as np
from sklearn.svm import SVR
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('aapl.csv')[:19]
```

```python
dates  = df['Date']
dates = [int(x.split('-')[0]) for x in dates.values]
dates = np.asarray(dates).reshape(-1,1)

prices = df['Open']
prices=np.asarray(prices.values).reshape(-1,1)
```

```python
plt.figure()
plt.scatter(dates,prices,label = 'Previous Data')
plt.legend();
```

```python
future_dates=np.append([29,30,31],dates).reshape(-1,1)
```

```python
svr_lin = SVR(kernel='linear')
svr_poly = SVR(kernel='poly',degree=2,C=10)
svr_rbf = SVR(kernel='rbf',gamma=0.01,C=10)

svr_lin.fit(dates,prices)
svr_poly.fit(dates,prices)
svr_rbf.fit(dates,prices)

plt.figure()

plt.plot(future_dates,svr_lin.predict(future_dates),label='Linear SVM')
plt.plot(future_dates,svr_poly.predict(future_dates),label='Polynmial SVM')
plt.plot(future_dates,svr_rbf.predict(future_dates),label='RBF SVM')
plt.scatter(dates,prices,label='Original Data',color='k')
plt.vlines(28,np.min(prices),np.max(prices)+5,alpha=0.4,label='predict')
plt.legend()
```
