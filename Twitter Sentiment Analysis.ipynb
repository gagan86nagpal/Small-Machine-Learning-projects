```python
inp = input('Enter the buzz word: ')
polarity_list = []
public_tweet  = api.search(inp)
for tweets in public_tweet:
   # print(tweets.text)
    analysis = TextBlob(tweets.text)
    polarity_list.append(analysis.sentiment.polarity)
polarity_np = np.asarray(polarity_list)
print('Polarity of {}: {}'.format(inp,np.mean(polarity_np)))
fig,ax = plt.subplots()
plt.bar([i for i in range(0,len(polarity_list))] , polarity_list,label  = 'Histogram')
for x,y in enumerate(polarity_list):
    x=x-0.35
    if(y>=0):
        y=y+0.02
    else:
        y=y-0.08
    ax.text(x,y,'{:.1f}'.format(polarity_list[int(x+1)]),fontsize=8)
    
plt.ylim(-1,1);
```
